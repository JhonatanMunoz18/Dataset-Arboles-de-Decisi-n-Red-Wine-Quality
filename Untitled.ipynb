### 3. Seleccionar las características más relevantes para entrenar el modelo utilizando selección de características.

```python
x= ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
    'pH', 'sulphates', 'alcohol']
y= ['quality']
```

Ácido Cítrico y pH (-0.5419): También hay una correlación negativa moderada, lo que sugiere que niveles más altos de ácido cítrico están asociados con un menor pH.
Densidad y Alcohol (-0.4962): Existe una correlación negativa moderada entre la densidad y el contenido de alcohol, indicando que un aumento en la densidad podría estar asociado con un menor contenido de alcohol.

Correlaciones Débiles:

La mayoría de las correlaciones entre las variables como cloruros, dióxido de azufre libre y total, y el contenido de alcohol son débiles, lo que sugiere que estas variables pueden no tener relaciones significativas entre sí.

### Interpretación General:
Las variables de acidez (fija y cítrica) están estrechamente relacionadas con el pH y la densidad, lo que podría indicar que la acidez y la densidad son factores importantes en la calidad del vino.
La presencia de azúcares residuales no parece tener una correlación significativa con las demás variables, lo que sugiere que su impacto puede ser independiente de las otras características del vino.
La correlación entre densidad y alcohol sugiere que el contenido de alcohol podría ser influenciado por la densidad del vino, posiblemente relacionado con el proceso de fermentación.

### 4. Dividir el dataset en Train y Test para evaluar correctamente el modelo. 


```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.2, random_state=13)
```

```python
tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit(X_train, y_train)
```

```python
plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)
```
